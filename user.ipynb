```python
from pptx import Presentation
from pptx.util import Inches
```

```python
df = pd.read_csv('user_data.csv')
```

```python
df.head()
```

```python
# Plot the data showing ages with the highest income
plt.figure(figsize=(10, 6))
sns.barplot(x='Age', y='Total_Income', data=df)
plt.title("Income by Age", fontsize=16)
plt.xticks(rotation=45)

'''
The graph shows that users aged 46 earned the highest income 
followed by users aged 37 and 45 respectively while users aged 51 
earned the lowest income.
'''
```

```python
# Show the gender distribution across spending categories.
plt.figure(figsize=(10, 6))
sns.barplot(x="Category", y="Amount", hue="Gender", data=df, palette="muted")
plt.title("Gender Distribution Across Spending Categories", fontsize=16)
plt.xlabel("Spending Categories", fontsize=12)
plt.ylabel("Amount Spent", fontsize=12)
plt.xticks(rotation=45)  # Rotate category labels for better readability
plt.legend(title='Gender')

'''
The graph shows that the male gender have a higher spending across all categories 
except school fees when compared with the female gender.
'''
```

```python
# Create a PowerPoint presentation
presentation = Presentation()
```

```python
# Add a slide to the presentation
slide_layout = presentation.slide_layouts[5]  # Use a blank slide layout
slide = presentation.slides.add_slide(slide_layout)
```

```python
# Define the image path
img_path = "gender_distribution.png"
img_path = "income_by_age.png"
```

```python
# Add the chart image to the slide
left = Inches(1)
top = Inches(1.5)
slide.shapes.add_picture(img_path, left, top, width=Inches(8.5), height=Inches(5))
```

```python
# Save the PowerPoint presentation
presentation.save("client_presentation.pptx")
print("PowerPoint presentation saved as 'client_presentation.pptx'")
```
